### 16. Convertir moneda de una denominacin a otra

Escriba un programa donde se le pida al usuario la cantidad de dinero en euros () y la
denominacin a la que desea cambiar. Para este caso utilice Dlar americano, Dlar
australiano, Dlar canadiense, Libra esterlina, Yuan chino, Rublo y Franco suizo. El
programa debe imprimir en pantalla la cantidad que recibir en la denominacin pedida


```python
def convertir_divisa(cantidad, divisa_destino):
    tasas_de_cambio = {
        "Dlar americano": 1.22,
        "Dlar australiano": 0.76,
        "Dlar canadiense": 0.79,
        "Libra esterlina": 1.41,
        "Yuan chino": 0.16,
        "Rublo": 0.014,
        "Franco suizo": 1.11
    }

    if divisa_destino in tasas_de_cambio:
        cantidad_convertida = cantidad * tasas_de_cambio[divisa_destino]
        return cantidad_convertida
    else:
        return "La denominacin de divisa especificada no es vlida."

cantidad_euros = float(input("Ingrese la cantidad de dinero en euros (): "))
denominacion_destino = input("Ingrese la denominacin a la que desea cambiar (Dlar americano, Dlar australiano, Dlar canadiense, Libra esterlina, Yuan chino, Rublo, Franco suizo): ")

resultado = convertir_divisa(cantidad_euros, denominacion_destino)
if isinstance(resultado, str):
    print(resultado)
else:
    print(f"Recibir {resultado} {denominacion_destino}.")
```

### 17. Vocal o consonante

En este ejercicio, crear un programa que lea una letra del alfabeto del usuario. Si el
usuario ingresa a, e, i, o o u entonces su programa debera mostrar un mensaje que
indica que la letra ingresada es una vocal. De lo contrario, su programa debera mostrar
un mensaje que indique que la letra es una consonante. Considerar solo minsculas

```python
def verificar_letra(letra):
    vocales = ['a', 'e', 'i', 'o', 'u']
    if letra in vocales:
        return "La letra ingresada es una vocal."
    else:
        return "La letra ingresada es una consonante."

letra = input("Ingrese una letra del alfabeto en minsculas: ")

if len(letra) == 1 and letra.isalpha():
    resultado = verificar_letra(letra)
    print(resultado)
else:
    print("Por favor, ingrese una nica letra del alfabeto en minsculas.")
```
