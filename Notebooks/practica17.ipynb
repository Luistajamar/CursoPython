```python
def persona_mas_joven(sara, sofia, ana):
    if sara < sofia and sara < ana:
        return "Sara es la más joven."
    elif sofia < sara and sofia < ana:
        return "Sofía es la más joven."
    elif ana < sara and ana < sofia:
        return "Ana es la más joven."
    else:
        return "Hay dos o más personas con la misma edad más joven."

sara = int(input("Ingrese la edad de Sara: "))
sofia = int(input("Ingrese la edad de Sofía: "))
ana = int(input("Ingrese la edad de Ana: "))

resultado = persona_mas_joven(sara, sofia, ana)
print(resultado)
```

### 16. Convertir moneda de una denominación a otra

Escriba un programa donde se le pida al usuario la cantidad de dinero en euros (€) y la
denominación a la que desea cambiar. Para este caso utilice Dólar americano, Dólar
australiano, Dólar canadiense, Libra esterlina, Yuan chino, Rublo y Franco suizo. El
programa debe imprimir en pantalla la cantidad que recibirá en la denominación pedida


```python
def convertir_divisa(cantidad, divisa_destino):
    tasas_de_cambio = {
        "Dólar americano": 1.22,
        "Dólar australiano": 0.76,
        "Dólar canadiense": 0.79,
        "Libra esterlina": 1.41,
        "Yuan chino": 0.16,
        "Rublo": 0.014,
        "Franco suizo": 1.11
    }

    if divisa_destino in tasas_de_cambio:
        cantidad_convertida = cantidad * tasas_de_cambio[divisa_destino]
        return cantidad_convertida
    else:
        return "La denominación de divisa especificada no es válida."

cantidad_euros = float(input("Ingrese la cantidad de dinero en euros (€): "))
denominacion_destino = input("Ingrese la denominación a la que desea cambiar (Dólar americano, Dólar australiano, Dólar canadiense, Libra esterlina, Yuan chino, Rublo, Franco suizo): ")

resultado = convertir_divisa(cantidad_euros, denominacion_destino)
if isinstance(resultado, str):
    print(resultado)
else:
    print(f"Recibirá {resultado} {denominacion_destino}.")
```

### 17. Vocal o consonante

En este ejercicio, creará un programa que lea una letra del alfabeto del usuario. Si el
usuario ingresa ‘a’, ‘e’, ‘i’, ‘o’ o ‘u’ entonces su programa debería mostrar un mensaje que
indica que la letra ingresada es una vocal. De lo contrario, su programa debería mostrar
un mensaje que indique que la letra es una consonante. Considerar solo minúsculas

```python
def verificar_letra(letra):
    vocales = ['a', 'e', 'i', 'o', 'u']
    if letra in vocales:
        return "La letra ingresada es una vocal."
    else:
        return "La letra ingresada es una consonante."

letra = input("Ingrese una letra del alfabeto en minúsculas: ")

if len(letra) == 1 and letra.isalpha():
    resultado = verificar_letra(letra)
    print(resultado)
else:
    print("Por favor, ingrese una única letra del alfabeto en minúsculas.")
```
