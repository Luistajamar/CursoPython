### 8 Indique si la siguiente afirmación es verdadera o falsa: 
Las cadenas en Python son iterables, troceables e
inmutables. Explicar en caso de que sea falsa

1. Iterables
Las cadenas en Python son iterables, lo que significa que puedes iterar sobre sus caracteres uno por uno utilizando bucles. Por ejemplo:

Este código imprimirá cada carácter de la cadena 'Hola' en una nueva línea.

```python
cadena = "Hola"
for caracter in cadena:
    print(caracter)
```

2. Troceables (subscriptables)
Las cadenas en Python son troceables, lo que significa que puedes acceder a subcadenas o caracteres individuales utilizando la notación de subíndice (slicing). Por ejemplo:

Puedes utilizar índices negativos para contar desde el final de la cadena y omitir el inicio o el final en el slicing para tomar desde el principio o hasta el final de la cadena.

```python
cadena = "Hola"
print(cadena[0])  # Imprime 'H'
print(cadena[1:3])  # Imprime 'ol'
```

3. Inmutables
Las cadenas en Python son inmutables, lo que significa que una vez creadas, no pueden ser modificadas. Cualquier operación que parezca modificar una cadena en realidad crea una nueva cadena. Por ejemplo:

```python
cadena = "Hola"
# Intentando modificar el primer carácter (esto producirá un error)
# cadena[0] = 'h'  # Error

# En su lugar, puedes crear una nueva cadena
nueva_cadena = 'h' + cadena[1:]
print(nueva_cadena)  # Imprime 'hola'
```

En resumen, las cadenas en Python cumplen con las propiedades de ser iterables, troceables e inmutables, lo que hace que la afirmación original sea correcta.
