### 1 Desarrolle un programa que le pregunte al usuario por la edad y que luego imprima en pantalla si dicho usuario es menor o mayor de edad.

```python
edad = int(input("Por favor, ingrese su edad: "))

if edad >= 18:
    print("Usted es mayor de edad.")
else:
    print("Usted es menor de edad.")
```

### 2 Escribir un programa que pida al usuario un número entero y que luego imprima en pantalla si dicho número es o no divisible entre 2

```python
numero = int(input("Por favor, ingrese un número entero: "))

if numero % 2 == 0:
    print("El número", numero, "es divisible entre 2.")
else:
    print("El número", numero, "no es divisible entre 2.")
```

### 3 Escriba un programa que pida al usuario dos números enteros y que calcule su división, indicando si la división es exacta o no

```python
numero1 = int(input("Por favor, ingrese el primer número entero: "))
numero2 = int(input("Por favor, ingrese el segundo número entero: "))

resultado = numero1 / numero2
if numero1 % numero2 == 0:
    print("La división es exacta.")
else:
    print("La división no es exacta.")

print("El resultado de la división de", numero1, "por", numero2, "es:", resultado)
```

## ACTIVIDAD 4

Dada una variable llamada “year “ con un valor entero, compruebe si dicho año es bisiesto o no lo es. Un año es bisiesto en el Calendario Gregorian, si es divisible entre 4 y no divisible entre 100, o bien si es divisible 

entre 400.

Ejemplo:

Entrada: 2008
Salida: Es un año bisiesto

Usar solo el condicional if (no usar funciones)



```python
year = 2008

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("El año", year, "es un año bisiesto.")
else:
    print("El año", year, "no es un año bisiesto.")
```

## ACTIVIDAD 5

Escriba un programa que pida al usuario su año de nacimiento, y que luego imprima en pantalla a que generación pertenece según la siguiente taxonomía



```python
anio_nacimiento = int(input("Por favor, ingrese su año de nacimiento: "))

if 1930 <= anio_nacimiento <= 1948:
    print("Usted pertenece a la Generación Silenciosa.")
elif 1949 <= anio_nacimiento <= 1968:
    print("Usted pertenece a la Generación Baby Boomers.")
elif 1969 <= anio_nacimiento <= 1980:
    print("Usted pertenece a la Generación X.")
elif 1981 <= anio_nacimiento <= 1993:
    print("Usted pertenece a la Generación Y (Millennials).")
elif 1994 <= anio_nacimiento <= 2010:
    print("Usted pertenece a la Generación Z.")
elif 2011 <= anio_nacimiento <= 2025:
    print("Usted pertenece a la Generación T (ALFA).")
else:
    print("Lo siento, no se puede determinar su generación con la información proporcionada.")
```
